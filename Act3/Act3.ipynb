```python
#Import the Tensorflow library, tf is  an alias or a short name  to make it easier  when  writing codes
import tensorflow as tf
#Import numpy
import numpy as np
#Import Keras, which is a high-level API inside TensorFlow
from tensorflow import keras
#Import the layers module from Keras.
from tensorflow.keras import layers
```

```python
#Create a  fully connected neural network layer with 10 neurons and using the ReLU activation

denseLayer = layers.Dense(units=10, activation='relu')
print(denseLayer)
```

```python
# Passing Data Through a Layer 
sampleInput = np.array([[1.0, 2.0, 3.0]])
output = denseLayer(sampleInput)
print(output)
```

```python
# Creating a Sequential Model.

model = keras.Sequential(
    [
        keras.Input(shape=(5,)),        # input layer
        layers.Dense(2, activation="relu"),
        layers.Dense(3, activation="relu"),
        layers.Dense(4),
       
    ]
)
print(model.summary())
```
